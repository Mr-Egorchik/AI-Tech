Работа с пропущенными значениями:

    удаление строк с пропущенными значениями;
    заполнение пропущенных значений средним значением по колонке.


```python
data.info()
```

```python
#Пропусков нет, нагенерируем их
import random
import numpy as np

rows = data.shape[0]
cols = data.shape[1]
types = data.dtypes

for i in range(1001):
    row = data.iloc[random.randint(0, rows-1)].copy()
    row.iloc[random.randint(1, cols-3)] = np.nan
    data = pd.concat([data, row.to_frame().T], ignore_index = True)

data = data.astype(types)
```

```python
data.info()
```

```python
#Удалим строки с пропусками в нечисловых столбцах
data = data.dropna(subset=["Apartment type", "Metro station", "Region"])
```

```python
data.info()
```

```python
#В числовых признаках заменим на среднее
cols_float = ["Minutes to metro", "Area", "Living area", "Kitchen area"]
cols_int = ["Number of rooms", "Floor", "Number of floors"]

data[cols_float] = data[cols_float].fillna(data[cols_float].mean())
data[cols_int] = data[cols_int].fillna(data[cols_int].mean())
data[cols_int] = data[cols_int].round(0)
```

```python
data.info()
```

Создание нового поля, вычисленного на основе значений других полей:

    через выражение на базе имеющихся колонок;
    через DataFrame.apply;
    через Series.apply.


```python
#Добавим цену за квадратный метр
data["Price per m^2"] = (data["Price"] / data["Area"]).round(2)
```

```python
#Добавим отношение жилой площади к общей площади
data["Living area(% of Area)"] = data.apply(lambda row: row["Living area"] / row.Area * 100, axis=1)
```

```python
#Добавим примерно расстояние до станции метро (при средней скорости человека в 5 км/ч)
data["Distance to metro (m)"] = data["Minutes to metro"].apply(lambda val: round(val / 60 * 5 * 1000))
```
